```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/cleaned_bank_data.csv')

df.head()
```

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[0, 29, 45, 60, 100],
    labels=['<30', '30-45', '46-60', '60+']
)

df['AgeGroup'].value_counts()
```

```python
df['CreditScoreBand'] = pd.cut(
    df['CreditScore'],
    bins=[0, 599, 749, 900],
    labels=['Low', 'Medium', 'High']
)

df['CreditScoreBand'].value_counts()
```

```python
df['TenureGroup'] = pd.cut(
    df['Tenure'],
    bins=[-1, 2, 7, 10],
    labels=['New', 'Mid-Term', 'Long-Term']
)

df['TenureGroup'].value_counts()
```

```python
def balance_segment(balance):

    if balance == 0:
        return 'Zero Balance'

    elif balance < 100000:
        return 'Low Balance'

    else:
        return 'High Balance'


df['BalanceSegment'] = df['Balance'].apply(balance_segment)

df['BalanceSegment'].value_counts()
```

```python
df.head()
```

```python
df.to_csv(
    '../data/processed/segmented_bank_data.csv',
    index=False
)

print("Segmented dataset saved successfully!")
```
